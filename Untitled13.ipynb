```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/students_Intellipaat.csv')
```

```python
print(df.head())
```

```python
print(df.isnull().sum())
df.duplicated().sum()
df.describe(include='all')
```

```python
import re

# If "Eemarks" column exists, merge it into Remarks
if "Eemarks" in df.columns:
    df["Remarks"] = df["Remarks"].astype(str) + df["Eemarks"].astype(str)
    df.drop(columns=["Eemarks"], inplace=True)


# Function to remove special characters
def clean_text(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

# Apply cleaning to Name and Remarks
df['Name'] = df['Name'].apply(clean_text)
df['Remarks'] = df['Remarks'].apply(clean_text)

#print(df[['Student_ID', 'Name', 'Remarks']])
print(df)
```

```python
# Calculate average sleep
avg_sleep = df["Sleep_Hours"].mean()
print("Overall Average Sleep Hours:", avg_sleep)
```

```python
# Filter condition
high_stress_sleep = df[(df["Sleep_Hours"] > avg_sleep) & (df["Stress_Level"] == "High")]

print(high_stress_sleep[["Student_ID", "Name", "Sleep_Hours", "Stress_Level"]])
```
